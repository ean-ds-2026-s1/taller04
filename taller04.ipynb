## La clase Cliente

En esta clase vamos a almacenar la información de un cliente. 

### Atributos

La clase `Cliente` será una **dataclass** y tendrá los siguientes atributos:

* `cedula`: un número entero que identifica de manera única
   cada cliente del banco.
* `nombres`: una cadena de caracteres (`String`) donde tenemos
  el nombre del cliente.
* `puntaje`: este número entero guarda el puntaje crediticio
  del cliente. 
* `pais`: aquí guardamos el identificador entero del país
  donde nació el cliente. Aquí usamos el número 1 para
  indicar que el país es Colombia, el número 2 es para
  Francia, el número 3 es para Ecuador y el número 4 es
  para España.
* `genero`: se refiere al sexo o género del cliente. Vamos
  a usar un número entero de la siguiente forma: 1 para los
  hombres y 2 para las mujeres.
* `edad`: la edad del cliente, un número entero.
* `año_abrio_cuenta`: un número entero que indica el año en que el cliente abrió su cuenta en el banco.
* `saldo`: un número entero que indica cuanto dinero tiene guardado el cliente en el banco.
* `tipo_cuenta`: el banco maneja varios tipos de producto. 
  Aquí indicamos cuál tipo de cuenta tiene el cliente. Este
  es un número entero que puede ser 1 para cuenta de ahorros,
  2 para cuenta corriente, 3 para cuenta especial y 4 para 
  créditos.
* `tiene_tarjeta`: un valor booleano que indica si el 
  cliente tiene tarjeta débito (`True`) o si no tiene la
  tarjeta (`False`).
* `activo`: un valor booleano que indica si el cliente 
  está activo en el banco (`True`) o si tiene mucho tiempo
  sin usar la cuenta, o sea, está inactivo (aquí es igual a `False`).
* `salario`: el dinero que recibe el cliente 
  mensualmente de su empresa. Es un valor entero.

### Métodos

Estos son algunos métodos que obtienen información del cliente y que están dentro de la **dataclass** `Cliente` que estamos construyendo:

- `dar_nombre_pais`: este método no recibe datos de entrada y en este caso va a retornar el nombre del país en 
    mayúsculas, teniendo en cuenta el código de este país. Así:
    si el `pais` es 1, deberá retornar `"COLOMBIA"`, y así
    sucesivamente. Retorne la palabra "ERROR" si el atributo
    `pais` no es ninguno de los 4 establecidos.
- `es_mujer`: un método que no recibe datos de entradas y retornará un valor `bool`, el cual será `True` si el cliente es mujer, es
    decir, el `genero` es 2 y `False` en caso contrario.
- `es_hombre`: un método que no recibe datos de entrada y que retornará `True` si el cliente es un hombre y `False` en caso contrario.
- `hombre_par`: método que no recibe datos de entrada y que retornará `True` si la cédula del cliente es un número
    positivo y par, es un hombre, tiene tarjeta y abrio su cuenta hace más de 30 años; retorne `False` en caso contrario.
- `puntaje_correcto`: que tampoco tiene datos de entrada y que
    retornará `true` si el puntaje crediticio es un número
    positivo de 3 dígitos y `False` en caso contrario.
- `es_europeo`: este método no recibe datos de entrada y retornará `True` si la persona nació en un país europeo (es decir, si nació en Francia o España) y `False` en caso contrario.  
- `seguro_social`: este método no recibe parámetros y retorna un número real que indica el valor a pagar por la seguridad social al gobierno. La regla para obtener este valor es la siguiente: para los      menores de edad, este valor es de cero pesos. Para los mayores de edad en 
    un país europeo (Francia o España), retorne el 30% del salario;
    para los otros países, retorne el 12% del salario.
- `es_apto`: permite saber si el cliente se queda en el banco. 
    Este método no recibe datos de entrada y retornará `True` si se trata de un cliente activo con una
    edad entre 30 y 40 años, que haya nacido en Francia o en Ecuador y que
    haya abierto la cuenta en el banco hace 1 año o másy tiene tarjeta débito. 
    En cualquier otro caso deberá retornarse `False`.
- `saldo_en_rojo`: este método no recibe datos de entrada y deberá retornar `True` cuando se
    tiene un cliente activo cuyo tipo de cuenta sea una cuenta corriente, no tiene tarjeta y el saldo sea negativo. En cualquier otro caso deberá retornarse `False`. 
- `aumentar_saldo`: este método recibe un solo dato de entrada,
    un número entero llamado `valor` y retorna un `bool` de 
    acuerdo con las siguientes condiciones:
    * Cuando el tipo sea cuenta de ahorros o cuenta corriente y el cliente está activo y es mayor de edad y el valor es positivo, se le
    suma al saldo de la cuenta (`saldo`) el parámetro `valor` y retornamos `True`
    * Para la cuentas corrientes o de ahorros y el valor es positivo y el cliente está activo pero es menor de edad, le sumamos al `saldo` la mitad del valor y también retornamos `True`
    * En cualquier otro caso, retornamos `False`.
- `disminuir_saldo`: este método recibe un solo dato de entrada, un número entero llamado `valor` y retorna un `bool` de acuerdo a las siguientes reglas:
   * Si el tipo es ahorros y el valor es positivo y el valor es inferior o igual al saldo y el cliente está activo, se le resta al atributo `saldo` el parámetro `valor` y retornamos `True`.
   * Si el tipo es cuenta corriente y el valor es positivo y el cliente está activo, entonces le restamos al atributo `saldo` el parámetro `valor` y retornamos `True`.
   * En cualquier otro caso, retornamos `False`

Los siguientes son los métodos de la **dataclass** `Banco`:

* `es_menor_de_edad`: método que no recibe ningún dato de entrada y retornará `True` cuando el
